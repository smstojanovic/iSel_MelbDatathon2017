# Get dataframe

```python
antiflu = feather.read_dataframe('..\\..\\Data\\Datathon_2017\\antiflu_sample_data')
```

```python
antiflu[['Patient_ID','lat','long']].head(2)
```

```python
grid_spacing = 300
```

```python
lat_grid = np.linspace(-45,-15,grid_spacing)
lon_grid = np.linspace(110,160,grid_spacing)
```

```python
def between(x,num1,num2):
    if (x < num2 and x >= num1):
        return(True)
    else:
        return(False)
```

```python
between(38,35,37)
```

```python
#lat_grid
np.sum(antiflu.apply(lambda x: between(x['lat'],-36,-35), axis = 1))
#antiflu[[between(antiflu['lat'],-37,-36)]]
```

```python
x = np.transpose(np.array(antiflu[['long']]))[0]
y = np.transpose(np.array(antiflu[['lat']]))[0]
```

```python
xv, yv = np.meshgrid(lat_grid[:(grid_spacing-1)], lon_grid[:(grid_spacing-1)])
```

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
H, xedges, yedges = np.histogram2d(x,y,bins=(lon_grid,lat_grid))
```

```python
H = H / np.sum(H)
```

```python
trace0 = Heatmap(z = H.T)
```

```python
 #iplot(Figure(data = [trace0]))
```

```python
H = H.reshape((grid_spacing-1)*(grid_spacing-1),1)
xv = xv.reshape((grid_spacing-1)*(grid_spacing-1),1)
yv = yv.reshape((grid_spacing-1)*(grid_spacing-1),1)
```

```python
df = np.concatenate((H,xv,yv), axis = 1)
```

```python
df = pd.DataFrame(df, columns=['density','lat','lon'])
```
